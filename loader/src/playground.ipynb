```python
import json
import pandas as pd
import random
terms_fp = './../data/terms.json'
annotations_fp = './../data/annotations.json'
human_fp = './../data/human.tsv'
```

```python
def get_pd_row(df, k):
    return dict(df.loc[k].dropna())

def denormalize_terms(df, terms):
    return [get_pd_row(df, term) for term in terms]

noterms = set()
def get_pd_row(df, k):
    try:
        return dict(df.loc[k].dropna())
    except KeyError:
        noterms.add(k)
        return None
```

```python
terms_df = pd.read_json(terms_fp)
terms_df = terms_df.set_index('ID', drop=False)
terms_df = terms_df.rename(columns={'ID': 'id', 'LABEL': 'label', 'hasOBONamespace':'aspect'})
json_chunk = terms_df.to_json(orient="records", default_handler=None)
json_str = json.loads(json_chunk)
terms_df

#write_to_json(json_str, ospath.join('out.json'))
```

```python


ann_df = pd.read_json(annotations_fp)
ann_df['term3'] = ann_df['term'].apply(lambda x: get_pd_row(terms_df, x))
ann_df['slim_terms2'] = ann_df['slim_terms'].apply(lambda x: denormalize_terms(terms_df, x))

json_chunk = ann_df.to_json(orient="records", default_handler=None)
json_str = json.loads(json_chunk)
ann_df
```

```python
filtered_terms = terms_df[terms_df['is_goslim']==True]
filtered_terms
```

```python
human_df = pd.read_csv(human_fp, sep='\t')
human_df = human_df.rename(columns={human_df.columns[0]:'gene'})
human_df = human_df[['gene']]

json_chunk = human_df.to_json(orient="records", default_handler=None)
json_str = json.loads(json_chunk)
human_df
```
